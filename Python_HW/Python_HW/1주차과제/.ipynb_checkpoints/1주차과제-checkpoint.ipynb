```python
# 파일 불러오기
fr=open("students.txt","r")
data=fr.read()
fr.close()

# 데이터 리스트(list)자료형으로 나누기 
import re
score=re.split('\t|\n',data)
    
# 0. 파일 생성하기 
student_list=[score[4*i] for i in range(0,5)]
student_list

name_dic={string:score[4*i+1]for i,string in enumerate(student_list)}
name_dic

midterm_dic={string:int(score[4*i+2])for i, string in enumerate(student_list)}
midterm_dic

Final_dic={string:int(score[4*i+3])for i, string in enumerate(student_list)}
Final_dic

Ave_dic={string:((int(score[4*i+2])+int(score[4*i+3]))/2)for i, 
         string in enumerate(student_list)}
Ave_dic

def grade(x):
   
    avg=Ave_dic[student_list[x]]
    ke=""
        
    if avg >= 90:
        ke+='A'
    elif avg >= 80:
        ke+='B'
    elif avg >= 70:
        ke+='C'
    elif avg >= 60:
        ke+='D'
    else : 
        ke+='F'
    return ke   

gra_dic={student_list[i]:grade(i) for i in range (0,5)}


# 1. show (전체 학생 정보 출력) 

def show():
    
    print("Student            Name   Midterm   Final   Average   Grade")
    print("-----------------------------------------------------------")
    for i in range (0,5):
        stu=student_list[i]
        print('%s%15s%10s%8s%10s%8s'%(student_list[i],name_dic[stu],midterm_dic[stu],Final_dic[stu], 
                                      Ave_dic[stu], gra_dic[stu]))

# 2. search (특정 학생 검색)

def search():    
    
    stu=input("Student ID: ")
    
    if stu in student_list:
        print("Student            Name   Midterm   Final   Average   Grade")
        print("-----------------------------------------------------------")
        print('%s%15s%10s%8s%10s%8s'%(student_list[i],name_dic[stu],midterm_dic[stu],Final_dic[stu], 
                                      Ave_dic[stu], gra_dic[stu]))
    else : 
        print ("NO SUCH PERSON.")

# 3. changescore (점수 수정)

def changescore():
    
    stu=input("Student ID: ")
    
    if stu in student_list: 
        test=input("Mid/Final? ")
        if test=="mid":
            ns=input("Input new score: ")
            if 0 <= int(ns) <=100:
                print("Student            Name   Midterm   Final   Average   Grade")
                print("-----------------------------------------------------------")
                print('%s%15s%10s%8s%10s%8s'%(stu,name_dic[stu],midterm_dic[stu],Final_dic[stu], 
                                              Ave_dic[stu], gra_dic[stu]))
                midterm_dic[stu]=int(ns)
                print("Score changed.")
                print('%s%15s%10s%8s%10s%8s'%(stu,name_dic[stu],midterm_dic[stu],Final_dic[stu], 
                                              Ave_dic[stu], gra_dic[stu]))
                
            else: 
                pass 
        elif test=="final":
            ns=input("Input new score: ")
            if 0 <= int(ns) <=100:
                print("Student            Name   Midterm   Final   Average   Grade")
                print("-----------------------------------------------------------")
                print('%s%15s%10s%8s%10s%8s'%(stu,name_dic[stu],midterm_dic[stu],Final_dic[stu], 
                                              Ave_dic[stu], gra_dic[stu]))
                Final_dic[stu]=int(ns)
                print("Score changed.")
                print('%s%15s%10s%8s%10s%8s'%(stu,name_dic[stu],midterm_dic[stu],Final_dic[stu], 
                                              Ave_dic[stu], gra_dic[stu]))
                
            else: 
                pass
        else :
            pass 
    else :
        print ("NO SUCH PERSON.")


# 4. add (학생 추가)

def add(): 
    stu=input("Student ID: ")
    
    if stu in student_list: 
        print("ALREADY EXISTS.")
        
    else: 
        
        na=input("Name: ")
        mid=input("Midterm Score: ")
        fin=input("Final Score: ")
        
        student_list.append("stu")
        name_dic[len(student_list)]=na
        midterm_dic['stu']=int(mid)
        Final_dic['stu']=int(fin)
        Ave_dic['stu']=(int(mid)+int(fin))/2
        gra_dic['stu']=grade((int(mid)+int(fin))/2)
        
        print("Student added.")

# 5. searchgrade (Grade 검색) 

def searchgrade(): 
    gr=input("Grade to search: ")
    if gr in gra_dic.values():
        a=[]
        for i in range(len(student_list)):
            if gra_dic[i]==gr:
                a.append(i)
            else:
                pass  
        return a
    else :
        print("NO RESULTS.") 
        
searchgrade()
        
# 6. REMOVE (특정 학생 삭제)   

#def REMOVE(): 
    #stu=input("Student ID: ")
    #if len()!==0:
        #if stu in : 
            
        #else: 
            #print("NO SUCH PERSON.")
    #else: 
        #print("List is empty.")
        
# 7. quit (종료)
        
#def quit(): 
    #a=input("Save data?[yes/no]: ")
    #if a == "yes":
        #return 
    #else : 
        #fa.write()
```
