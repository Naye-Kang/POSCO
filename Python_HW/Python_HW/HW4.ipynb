```python
#실습2 

File_name=input("Enter a file name: ")
import os 
if os.path.exists(File_name):
    f=open(File_name,"r")
    data=f.read()
    print(data.upper())
    f.close()
else : print("파일이 존재하지 않는다")
```

```python
#실습3 

import sys
print(sys.argv)  
fr=open(sys.argv[1],"r")
fw=open(sys.argv[2],"w")
for line in fr: 
    fw.write(line)
fr.close()
fw.close()
```
